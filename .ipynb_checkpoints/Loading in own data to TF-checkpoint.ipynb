```python
import numpy as np # to do array operations
import matplotlib.pyplot as plt # to show the image
import os # to iterate through directories and join paths
import cv2 # to do image operations

DATADIR = "/Users/raghavsharma/Downloads/kagglecatsanddogs_3367a/PetImages/"
CATEGORIES = ["Dog", "Cat"]

for category in CATEGORIES:
    path = os.path.join(DATADIR, category)  # path to cats or dogs dir
    
    # iterate through these images
    for img in os.listdir(path):
        # convert to grayscale to 1/3 the size also color not imp in this task
        img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
        plt.imshow(img_array, cmap="gray")
        plt.show()
        break
    break
```

```python
print(img_array)
```

```python
# resize every image to the same size
IMG_SIZE = 100

new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
plt.imshow(new_array, cmap="gray")
plt.show()
```

```python
# training data
training_data = []

def create_training_data():
    for category in CATEGORIES:
        path = os.path.join(DATADIR, category)  # path to cats or dogs dir
        class_num = CATEGORIES.index(category)  # 0 for dog and 1 for cat
        # iterate through these images
        for img in os.listdir(path):
            try:
                # convert to grayscale to 1/3 the size also color not imp in this task
                img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
                new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
                training_data.append([new_array, class_num])
            except Exception as e:
                pass

create_training_data()      
```

```python
print(len(training_data))
```

```python
# shuffle the data
import random
random.shuffle(training_data)
```

```python
for sample in training_data[:10]:
    print(sample[1])
```

```python
X = []
y = []

for features, label in training_data:
    X.append(features)
    y.append(label)
    
X = np.array(X).reshape(-1, IMG_SIZE, IMG_SIZE, 1)
```

```python
# save our data for tweeking later and not build it again
import pickle

pickle_out = open("X.pickle", "wb")
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open("y.pickle", "wb")
pickle.dump(y, pickle_out)
pickle_out.close()
```
